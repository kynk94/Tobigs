```python
tokenize = kkma.nouns(sample)
tokenize
```

```python
CBOW_Model = Word2Vec(tokenize, window = 5, size = 10, min_count = 3, sg = 0)
CBOW_Model
```

```python
Skip_Gram_Model = Word2Vec(tokenize, window = 5, size = 10, min_count = 3, sg = 1)
Skip_Gram_Model
```

```python
CBOW_Model_Result = dict(zip(CBOW_Model.wv.index2word, CBOW_Model.wv.vectors))
CBOW_Model_Result
```

```python
Skip_Gram_Model_Result = dict(zip(Skip_Gram_Model.wv.index2word, Skip_Gram_Model.wv.vectors))
Skip_Gram_Model_Result
```

```python
CBOW_Model.wv.most_similar("가")
```

```python
Skip_Gram_Model.wv.most_similar("가")
```
